```python
from pennsieve import Pennsieve
import matplotlib.pyplot as plt
import pandas as pd
```

```python
p = Pennsieve()
p.use_dataset('N:dataset:35881ab1-e289-42b6-a21c-7ded18b54be7')
```

### Get channel objects for package
Using the `getChannels` method, we get channel objects, their names, sampling rates etc.

```python
channels = p.timeseries.getChannels(p.dataset, 'N:package:8c223e2a-f259-4869-b888-fd29ec2a8ee4', True)
print(channels[0])
```

### Get range of data as for list of channels
Using the `getRangeForChannels`, we get the data for a set of channels as a Pandas Dataframe.

We are getting 2 channels and 2 seconds of data at 200Hz

```python
res = p.timeseries.getRangeForChannels(p.dataset, 'N:package:8c223e2a-f259-4869-b888-fd29ec2a8ee4', [channels[1].id, channels[2].id],499,501, False, True)
```

### Simple plot of the data

```python
res.plot()
plt.show()
```

Using the `getRangeForChannels`, we get the data for a set of channels as a Pandas Dataframe.

We are getting all channels (11) and 100 seconds of data at 200Hz
